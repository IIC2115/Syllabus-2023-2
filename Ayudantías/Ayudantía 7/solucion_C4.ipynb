```python
import geopandas as gpd
```

```python
areas_urbanas = gpd.read_file('data/Areas Urbanas/areas_urbanas.shp')
distritos = gpd.read_file('data/Distritos Censales/Distritos Censales RM.shp')
lineas_metro = gpd.read_file('data/Metro 2020/Lineas_2020/Lineas_2020.shp')
estaciones_metro = gpd.read_file('data/Metro 2020/Estaciones_2020/Estaciones_2020.shp')
```

```python
santiago_urbano = distritos.loc[distritos['NOM_REG']=='Región Metropolitana De Santiago']
print(santiago_urbano['NOM_PROV'].value_counts())
santiago_urbano.plot()
```

```python
santiago_urbano = santiago_urbano.loc[santiago_urbano['NOM_PROV']=='Santiago']
santiago_urbano.plot()
```

```python
santiago_urbano.plot(column = 'ADIMARK_T', figsize = (10,10))
```

```python
# Vamos a proponer una forma de nivel economico
personas = santiago_urbano['E'] * 0.1 + santiago_urbano['D'] * 0.25 + santiago_urbano['C3'] * 0.5 + santiago_urbano['C2'] * 0.65 + santiago_urbano['ABC1'] * 0.8
santiago_urbano['Nivel Economico'] = santiago_urbano['1_UF_M2'] * personas - santiago_urbano['Densidad']
santiago_urbano.plot(column = 'Nivel Economico', figsize = (10,10))
```

# Misión 2

```python
estaciones_metro_utm = estaciones_metro.to_crs(32719)
buffer_metro = estaciones_metro_utm.copy()
buffer_metro.geometry = buffer_metro.buffer(100)
buffer_metro.plot()
```

```python
distritos_intersecciones = gpd.overlay(distritos, buffer_metro, how='intersection')
distritos_filtrados = distritos_intersecciones.groupby('ESTACION').filter(lambda x: len(x) > 2)
print(distritos_filtrados['ESTACION'].unique())
```

# Misión 3

```python
estaciones_distritos = gpd.sjoin(estaciones_metro, distritos, how="inner", op="within")
estaciones_distritos['NOM_COM'].value_counts()
```

```python
estaciones_por_distrito_group = estaciones_distritos.groupby('NOM_COM').size().reset_index(name='cantidad_estaciones')
estaciones_por_distrito_group
```

```python
estaciones_por_distrito = estaciones_por_distrito_group.sort_values(by='cantidad_estaciones', ascending=False)

estaciones_por_distrito.head()
```

```python
distritos_con_estaciones = distritos.merge(estaciones_por_distrito_group, on='NOM_COM', how='left')
distritos_con_estaciones.plot(column = 'cantidad_estaciones', figsize = (10,10))
```
