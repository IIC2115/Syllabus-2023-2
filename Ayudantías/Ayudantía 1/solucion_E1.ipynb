```python
import random
from abc import ABCMeta, abstractmethod
```

# Definir clases

## Local

```python
class Local():
    def __init__(self, filas, p, aforo):
        self.n_filas = filas
        self.filas = []
        self.probabilidad = p
        self.aforo = aforo
        self.gente_total = 0
        self.fila_espera = []
        self.inicializar_filas()

    def inicializar_filas(self):
        for i in range(self.n_filas):
            self.filas.append([])

    def llegada_votante(self):
        return random.choices([True, False], weights=[self.probabilidad, 1 - self.probabilidad], k=1)[0]

    def incorporar_votante(self, votante):
        if self.gente_total < self.aforo:
            self.filas[votante.mesa].append(votante)
            self.gente_total += 1
            print(f"Votante N°{votante.id} ingresado en la mesa {votante.mesa}")
        else:
            print(f"Votante N°{votante.id} ingresado en la fila de espera")
            self.fila_espera.append(votante)


    def avanzar_tiempo_votacion(self):
        for i in range(n_filas):
            if len(self.filas[i]) > 1:
                termino = self.filas[i][0].votar()
                if termino:
                    self.gente_total -= 1
                    antiguo_votante = self.filas[i].pop(0)
                    if len(self.fila_espera) > 0:
                        nuevo_votante = self.fila_espera.pop(0)
                        self.incorporar_votante(nuevo_votante)

    def __str__(self):
        str_final = ""
        for i in range(len(self.filas)):
            str_final += "--------------------------"
            str_final += f"Lista N° {i}\n"
            for votante in self.filas[i]:
                str_final += f"Votante N° {votante.id}\n"
        return str_final
```

## Personas Votantes

```python
class Persona(metaclass=ABCMeta):
    def __init__(self, edad):
        self.edad = edad
        self.tiempo_espera = 10

    @abstractmethod
    def seleccion_mesa(self, filas):
        pass

    @abstractmethod
    def votar(self):
        pass

    @abstractmethod
    def __str__(self):
        pass

class VotanteJoven(Persona):
    def __init__(self, edad, id_votante):
        super().__init__(edad)
        self.id = id_votante
        self.mesa = None

    def seleccion_mesa(self, filas):
        mesa = random.randint(0, filas - 1)
        self.mesa = mesa
        return mesa

    def votar(self):
        self.tiempo_espera -= 5
        if self.tiempo_espera < 0:
            print(f"Voto emitido\n{self}")
            return True
        else:
            return False

    def __str__(self):
        return f"Votante N°{self.id}\nEdad: {self.edad}\nMesa: {self.mesa}"

class VotanteAdulto(Persona):
    def __init__(self, edad, id_votante):
        super().__init__(edad)
        self.id = id_votante
        self.mesa = None

    def seleccion_mesa(self, filas):
        mesa = random.randint(0, filas - 1)
        self.mesa = mesa
        return mesa

    def votar(self):
        self.tiempo_espera -= 3
        if self.tiempo_espera < 0:
            print(f"Voto emitido\n{self}")
            return True
        else:
            return False

    def __str__(self):
        return f"Votante N°{self.id}\nEdad: {self.edad}\nMesa: {self.mesa}"

class VotanteMayor(Persona):
    def __init__(self, edad, id_votante):
        super().__init__(edad)
        self.id = id_votante
        self.mesa = None

    def seleccion_mesa(self, filas):
        mesa = random.randint(0, filas - 1)
        self.mesa = mesa
        return mesa

    def votar(self):
        self.tiempo_espera -= 1
        if self.tiempo_espera < 0:
            print(f"Voto emitido\n{self}")
            return True
        else:
            return False

    def __str__(self):
        return f"Votante N°{self.id}\nEdad: {self.edad}\nMesa: {self.mesa}"
```

# Simulación

```python
# creación de variables y objetos relevantes
aforo = 20
p = 0.75
n_filas = 5

local_votacion = Local(n_filas, p, aforo)
n_filas = 5
tiempo_a_simular = 100

id_global = 0

for t in range(tiempo_a_simular):
    llegada_nuevo_votante = local_votacion.llegada_votante()

    if llegada_nuevo_votante:
        # Entonces creamos un nuevo votante que se incorporará
        edad_nuevo_votante = random.randint(18, 100)
        if edad_nuevo_votante < 30:
            nuevo_votante = VotanteJoven(edad_nuevo_votante, id_global)
        elif edad_nuevo_votante < 60:
            nuevo_votante = VotanteAdulto(edad_nuevo_votante, id_global)
        else:
            nuevo_votante = VotanteMayor(edad_nuevo_votante, id_global)
        id_global += 1

        # Agregamos el nuevo votante
        nuevo_votante.seleccion_mesa(n_filas)
        local_votacion.incorporar_votante(nuevo_votante)

    # Realizamos las acciones que tienen que ver con el tiempo en base al local
    local_votacion.avanzar_tiempo_votacion()

print(local_votacion)
```
