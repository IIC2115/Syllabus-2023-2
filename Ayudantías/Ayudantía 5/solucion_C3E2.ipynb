```python
label_encoder = LabelEncoder()
```

```python
df_species = df.dropna(subset=['species'])
print(df_species['species'].value_counts())
array_species = label_encoder.fit_transform(df_species['species'])
series_species = pd.Series(array_species, name='species', index=df_species.index)
df_uso.update(series_species)
```

```python
df_uso.head()
```

```python
df_sex = df.dropna(subset=['sex'])
print(df_sex['sex'].value_counts())
df_sex = df_sex[df_sex != '.']
print(df_sex['sex'].value_counts())
array_sex = label_encoder.fit_transform(df_sex['sex'])
series_sex = pd.Series(array_sex, name='sex', index=df_sex.index)
df_uso.update(series_sex)
```

```python
df_island = df.dropna(subset=['island'])
print(df_island['island'].value_counts())
array_island = label_encoder.fit_transform(df_island['island'])
series_island = pd.Series(array_island, name='island', index=df_island.index)
df_uso.update(series_island)
```

```python
df_uso.head()
```

```python
scaler = StandardScaler()
columnas = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']
df_scalars = df.dropna(subset=columnas)
vectors_scalars = scaler.fit_transform(df_scalars[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']])
df_scalars = pd.DataFrame(vectors_scalars, columns=columnas, index=df_scalars.index)
df_uso.update(df_scalars)
df_scalars.head()
```

# Misión 2

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(df_limpio_uso.drop(['species'], axis=1), df_limpio_uso['species'], test_size=0.3)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo:", accuracy)
```
